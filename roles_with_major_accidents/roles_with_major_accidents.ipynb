```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
```

```python
df = pd.read_csv("d://Mining Sector//files//mining_company_data.csv")
df["HireDate"]  = pd.to_datetime(df["HireDate"],  errors="coerce")
df["ShiftDate"] = pd.to_datetime(df["ShiftDate"], errors="coerce")
 
sns.set_theme(style="whitegrid", palette="muted")
plt.rcParams.update({"figure.dpi": 130, "axes.titlesize": 13, "axes.labelsize": 11})
```

```python
df['IncidentShift'] = df['ShiftDate'].dt.strftime('%Y-%m-%d')
df['IncidentShift'] = pd.to_datetime(df['IncidentShift'], errors='coerce')
df['IncidentShift'] 
```

```python
major_acc = df[df["Incident"] == "Major Accident"]
role_acc  = major_acc["Role"].value_counts()
print(f"Role with most Major Accidents: {role_acc.idxmax()} ({role_acc.max()})")
```

```python
fig, ax = plt.subplots(figsize=(9, 4))
role_acc.plot(kind="bar", ax=ax, color=sns.color_palette("OrRd", len(role_acc)), edgecolor="white")
ax.set_title("Major Accident Incidents by Role")
ax.set_xlabel("Role")
ax.set_ylabel("Count")
ax.set_xticklabels(ax.get_xticklabels(), rotation=30, ha="right")
plt.tight_layout()
```
